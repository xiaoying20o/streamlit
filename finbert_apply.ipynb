```python
import pandas as pd
filepath = '/content/drive/MyDrive/cyber risk/hurricane_call_output.csv'
df = pd.read_csv(filepath)
```

```python
from transformers import pipeline
import pandas as pd

#load
finbert = pipeline("sentiment-analysis", model="ProsusAI/finbert")
df['hurricane_context'] = df['hurricane_context'].astype(str)
#apply model
def analyze_context(text):
    try:
        result = finbert(text[:512])[0]  # maximum length
        return pd.Series([result['label'], result['score']])
    except:
        return pd.Series(['error', 0.0])
df[['label', 'score']] = df['hurricane_context'].apply(analyze_context)
```
