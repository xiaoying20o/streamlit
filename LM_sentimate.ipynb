```python
import pysentiment2 as ps
lm = ps.LM()
def apply_lm(text):
    tokens = lm.tokenize(str(text)) # Ensure text is treated as string
    score = lm.get_score(tokens)
    return pd.Series([
        score.get('Positive', 0),
        score.get('Negative', 0),
        score.get('Polarity', 0),
    ])

df[['lm_pos', 'lm_neg', 'lm_polarity']] = df['hurricane_context'].apply(apply_lm)
```
